```python
import tensorflow as tf

class Clasificador:

    """
    Esta clase permite clasificar un texto o una lista de textos

    """

    def __init__(self, model, tokenizer, max_length=512):
        self.model = model
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __call__(self, text):
        """
        Esta función permite clasificar un texto o una lista de textos

        """
        if isinstance(text, str):
            text = [text]

        inputs = self.tokenizer(text, padding=True, truncation=True, max_length=self.max_length, return_tensors="tf")
        outputs = self.model(inputs)
        probs = tf.nn.softmax(outputs.logits, axis=-1)
        preds = tf.math.argmax(probs, axis=-1)
        preds = [self.model.config.id2label[pred] for pred in preds.numpy()]

        return preds
    

## prueba de la clase

clasificador = Clasificador(model_neg, tokenizer_neg)

clasificador(df.comentario.sample(10).tolist())
```
